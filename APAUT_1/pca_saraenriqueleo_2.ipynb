### b) Perform a PCA analysis over the data
El Análisis de Componentes Principales (PCA) se emplea para reducir la dimensionalidad de un conjunto de datos manteniendo la mayor cantidad de información posible. 

- En primer lugar, se identifican las direcciones (componentes principales) en las que los datos presentan mayor varianza. Estas direcciones son combinaciones lineales de las variables originales. 

- A continuación, se seleccionan los componentes principales que capturan la mayor parte de la varianza en los datos. 

- Por último, los datos originales se proyectan sobre estos componentes principales, creando un nuevo espacio con menos dimensiones pero que retiene las características esenciales.

La normalización de los datos es esencial antes de realizar un PCA, ya que si las variables originales tienen diferentes escalas, aquellas con valores más grandes dominarán el cálculo de la varianza, y las variables con valores más pequeños podrían perder importancia. Al estandarizar los datos, se eliminan las unidades y se pone a todas las características en la misma escala, permitiendo que el PCA capture patrones inherentes en los datos, no simplemente la influencia de magnitudes.

Con el fin de demostrar la necesidad de normalizar los datos, se hará en paralelo el análisis de componentes principales para los datos sin normalizar, discutiéndose el efecto que tiene dicha normalización.

Cálculo del número de componentes necesarias sin normalizar

```python
# Normalizamos los datos
data = wdbc.copy()
means = stats.loc['mean'] # Almacenamos las medias
stds = stats.loc['std'] # Almacenamos las desviaciones estándar
for column in range(2, data.shape[1]):
    data.iloc[:,column]=(data.iloc[:,column]-means[column])/stds[column] # Restamos la media y dividimos por la desviación estándar

X = data.values[:,2:].astype(float) # no nos interesa el identificador ni la etiqueta, guardamos los datos normalizados en X
X_notnormalized = wdbc.values[:,2:].astype(float) #guardamos los datos sin normalizar en X_notnormalized

data.head()
```

A continuación, se lleva a cabo el PCA para los datos normalizados, que será el que se utilice posteriormente para la representación de los dos componentes que capturan la mayoría de la varianza entre las muestras.

```python
from sklearn.decomposition import PCA

X_r= PCA(n_components=2).fit_transform(X)
```

También se realiza un PCA en el que se obtienen todos los componentes posibles a través de las funciones PCA().fit(). De esta forma se puede obtener un gráfico en el que se representa la varianza explicada por cada componente de forma acumulativa.

Primero, se realiza este procedimiento en los datos normalizados:

```python
X_pca= PCA().fit(X)

plt.plot(np.cumsum(X_pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

A continuación, se repite la representación con los datos sin normalizar.

```python
X_pca_notn= PCA().fit(X_notnormalized)

plt.plot(np.cumsum(X_pca_notn.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

Al comparar ambos gráficos, se observa que, para llegar al mismo valor de varianza explicada, es necesario un número mayor de componentes en los datos normalizados que en los datos sin normalizar. Esto se debe a que aquellos datos que presenten una varianza muy elevada (por tener valores en rangos superiores a otros) van a tener un mayor peso a la hora de calcular las varianzas explicadas por cada componente. 

A través de la normalización, evitamos el sesgo provocado porque dichos valores tengan una influencia superior en el cálculo de los componentes. De esta forma, los componentes principales reflejan las direcciones de varianza máxima sin estar afectados por las diferencias en los rangos de los valores de expresión para distintos genes. 

### c) Project the data into the PC space;

```python
plt.figure()
comp1=0; #first component to visualize, you can modify it
comp2=1; #second component to visualize, you can modify it

# plot the two components selected above for both malign and benign tumors
plt.scatter(X_r[y == 0, comp1], X_r[y == 0, comp2], color='r', alpha=.7, lw=1,
                label='malign')

plt.scatter(X_r[y == 1, comp1], X_r[y == 1, comp2], color='g', alpha=.1, lw=1,
                label='benign')

plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA WDBC dataset')


plt.show()
```

### d) Apply both k-means and GMM over the projected data;


```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
cluster = y & y_kmeans


plt.scatter(X_r[cluster == 1, comp1], X_r[cluster == 1, comp2], color='b', alpha=1, lw=1,
                label='malign')

plt.scatter(X_r[cluster == 1, comp1], X_r [cluster == 1, comp2], color='b', alpha=.1, lw=1,
                label='beign')

plt.scatter(X_r[y == 0, comp1], X_r[y == 0, comp2], color='r', alpha=.1, lw=1,
                label='malign')

plt.scatter(X_r[y == 1, comp1], X_r [y == 1, comp2], color='g', alpha=.1, lw=1,
                label='beign')

target_names_2 = np.array([('M'), ('B')], dtype = np.dtype('U10'))

data['cluster'] = target_names_2[y_kmeans]

plt.show()
```

```python
from sklearn.mixture import GaussianMixture as GMM

gmm = GMM(n_components=2).fit(X)
y_gmm = gmm.predict(X)

plt.scatter(X_r[y_gmm == 0, comp1], X_r[y_gmm == 0, comp2], color='r', alpha=.1, lw=1,
                label='malign')

plt.scatter(X_r[y_gmm == 1, comp1], X_r[y_gmm == 1, comp2], color='g', alpha=.1, lw=1,
                label='benign')
plt.show()
```
